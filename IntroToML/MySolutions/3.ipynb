```python

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal
import math as math
# Define mean and covariance matrix
u1 = [1, 2]
u2 = [0,-1]
cov1 = [[1, 0], [0, 2]]  # Covariance matrix
cov2 = [[2, 0], [0, 2]]
# Create a grid of points
N = 1000
X = np.linspace(-3, 5, N)
Y = np.linspace(-4, 5, N)
X, Y = np.meshgrid(X, Y)

# Create the multivariate normal object
N1 = multivariate_normal(u1,cov1)
N2 = multivariate_normal(u2,cov2)
# Evaluate the probability density function (PDF) at each point
pos = np.dstack((X, Y))  # Combine X and Y into a 3D array
Z1 = N1.pdf(pos)  # Evaluate PDF at all points
Z2 = N2.pdf(pos)
# Plot the results
M=200
#decision boundary
X1=np.linspace(-3,5,M)
X12=np.zeros(M)
for i in range(M):
  X12[i]=X1[i]* X1[i]
Y1=1/6*(X12)-2/3*X1+(5/6-1/3*math.log(3/math.sqrt(2)))*np.ones(M)
plt.plot(X1,Y1,color='b')
plt.contour(X, Y, Z1, colors='g')  # Use contour for lines
plt.contour(X, Y ,Z2, colors='r')  # Use contour for lines
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Bayes Classifier")
plt.show()
```
