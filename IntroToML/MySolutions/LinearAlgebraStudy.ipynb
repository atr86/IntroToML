Driver Function

```python
import sympy as sp

def gaussian_elimination(matrix):
    """
    Performs Gaussian elimination on a matrix using SymPy.

    Args:
      matrix: A SymPy Matrix object.

    Returns:
      A SymPy Matrix object in reduced row-echelon form.
    """
    m = matrix.rows
    n = matrix.cols

    for i in range(m):
        #Find the pivot
        pivot_row = i
        for k in range(i + 1, m):
            if abs(matrix[k, i]) > abs(matrix[pivot_row, i]):
                pivot_row = k

        #Swap rows if necessary
        if pivot_row != i:
          matrix.row_swap(i, pivot_row)

        #Eliminate entries below the pivot
        pivot = matrix[i, i]
        if pivot != 0:
          for k in range(i + 1, m):
              factor = matrix[k, i] / pivot
              for j in range(i, n):
                  matrix[k, j] -= factor * matrix[i, j]
    return matrix

# Example usage:
matrix = sp.Matrix([[2, 1, -1],
                    [-3, -1, 2],
                    [-2, 1, 2]])
print("Original Matrix:")
display(matrix)

reduced_matrix = gaussian_elimination(matrix.copy())
print("Matrix after Gaussian Elimination:")
display(reduced_matrix)

rref_matrix, _ = matrix.rref()
print("Matrix after Reduced Row Echelon Form using SymPy's rref():")
display(rref_matrix)
```
