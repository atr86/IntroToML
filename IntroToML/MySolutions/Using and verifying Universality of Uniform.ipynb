```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

l= np.random.random_sample(size=10000)
plt.title("Sample randomly so is a unif(0,1) distribution(in such way sampled by system)")
plt.hist(l,100)
plt.show()
```

```python
cdf_inv= stats.norm.ppf(l)
plt.title("By inverse cdf method, samping normal(0,1) from unif(0,1) ( Using Universility principle )")
plt.hist(cdf_inv,100)
plt.show()
```

```python
cdf=stats.norm.cdf(cdf_inv)
plt.title("Using normal(0,1), we calculate cdf of samples from normal(0,1) and sample from it (to verify universality of uniform)")
plt.hist(cdf,100)
plt.show()
print(np.mean(cdf))
print(np.var(cdf))
```
