```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

u= np.random.uniform(0,1,size=10000)
plt.title("Sample randomly so is a unif(0,1) distribution(in such way sampled by system)")
plt.hist(u,100)
plt.show()
```

```python
def FinvX(list_in):
  r=[]
  for i in list_in:
    r.append(1-np.sqrt(1-i))

  return r
```

```python
X=FinvX(u)
Y=np.random.uniform(0,4*(1-np.array(X)),10000)

plt.title("x+y/4<=1,x>=0,y>=0- JOINT DISTRIBUTION PLOT")
plt.scatter(X,Y,s=1)
plt.show()
```
