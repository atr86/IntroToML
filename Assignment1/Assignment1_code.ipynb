```python
import numpy as np
import matplotlib.pyplot as plt
```

# Q2 : using uniform to sample from Bern(p)

```python
# let 
p=0.4
N = 20000
u = np.random.uniform(0, 1, N)
sample_bernoulli = (u >= 1-p).astype('int')
N_1s = np.sum(sample_bernoulli)
N_0s = N - N_1s
plt.bar(["X=0", "X=1"], [N_0s/N, N_1s/N])
plt.ylabel("P(X)")
plt.title("generating bernoulli from uniform")
```

# Q3: using uniform to sample from  x+y/4<= 1, x>=0; y>=0

```python
# sample x
N = 2000
u = np.random.uniform(0, 1, N)
x = 1 - np.sqrt(u)

y = np.random.uniform(0, 4*(1-x), N)
plt.xlim([0, 4])
plt.scatter(x, y, s=1)
plt.plot([0, 1], [4, 0], c='r')
plt.title("Scatter: x+y<= 1, x>=0; y>=0")
plt.xlabel('x')
plt.ylabel('y')
```

# Q5: finding values of Z

```python
from scipy.stats import norm
#z-score = 8 sqrt(40)/5
p_tail = 1 - norm.cdf(8*np.sqrt(40.0)/5.0)
print(p_tail)
```
