```python
import numpy as np
import sympy as sp
from sympy import *
init_printing(use_unicode=True)
```

# **We find A 's row space and Null Space**

```python
arr=[[1,2,0,2,1],[3,6,1,9,6],[2,4,1,7,5]]
#print(arr)
m=len(arr)
n=len(arr[0])
'''for i in range(m):
    for j in range(n):
        print(arr[i][j],end=" ")
    print()
'''
M = sp.Matrix(arr)
display(M)
M=M.rref()[0]
print("RREF Form of the Matrix")
display(M)
```

```python
zero=0
one=0
r=0
a=np.array(M)
b=[0 for _ in range(n)]
for j in range(n):
  zero=0
  one=0
  for i in range(m):
       if a[i][j]==0:
         zero=zero+1
       if a[i][j]==1:
         one=one+1
  if one==1 and zero==m-1:
     b[j]=1
     r+=1
  else:
     b[j]=0
#print(b) WHICH COLUMNS CONTAIN PIVOT AND WHICH DOES NOT
print("Rank is",r)
```

```python
count=0
print("The Row Spaces are ")
for i in range(r):
    print(arr[i])
```

```python
k=0# gets to next column where F is there
l=0# gets to next row where F is there
N=[[0 for _ in range(n-r)] for _ in range(n)] #NULL SPACE
for j in range(n-r):
  while b[k]!=0:
        k+=1
  for i in range(n):
      if b[i]==1:
            N[i][j]=-a[l][k]
            l+=1
      else:
            if (i==k):
               N[i][j]=1
            else:
               N[i][j]=0
  l=0
  k=k+1
print("The Null Space is(Each Column is a basis vector)")
display(Matrix(N))
```

# **Using the same code, but now A transpose, we find Column space and Null space of A transpose**




```python
arr=np.transpose([[1,2,0,2,1],[3,6,1,9,6],[2,4,1,7,5]])
m=len(arr)
n=len(arr[0])
M = sp.Matrix(arr)
#display(M)
M=M.rref()[0]
#print("RREF Form of the Matrix")
display(M)
```

```python
zero=0
one=0
r=0
a=np.array(M)
b=[0 for _ in range(n)]
for j in range(n):
  zero=0
  one=0
  for i in range(m):
       if a[i][j]==0:
         zero=zero+1
       if a[i][j]==1:
         one=one+1
  if one==1 and zero==m-1:
     b[j]=1
     r+=1
  else:
     b[j]=0
#print(b)# WHICH COLUMNS CONTAIN PIVOT AND WHICH DOES NOT
print("Rank is",r)
```

```python
count=0
print("The Column Spaces are ")
for i in range(r):
    print(arr[i])
```

```python
k=0# gets to next column where F is there
l=0# gets to next row where F is there
N=[[0 for _ in range(n-r)] for _ in range(n)] #NULL SPACE
for j in range(n-r):
  while b[k]!=0:
        k+=1
  for i in range(n):
      if b[i]==1:
            N[i][j]=-a[l][k]
            l+=1
      else:
            if (i==k):
               N[i][j]=1
            else:
               N[i][j]=0
  l=0
  k=k+1
print("The Null Space of Matrix Transposed is(Each Column is a basis vector)")
display(Matrix(N))
```
