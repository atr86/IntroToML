```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample x
N = 20000
u = np.random.uniform(0, 1, N)
x = 1 - np.sqrt(u)

y = np.random.uniform(0, 1-x, N)

plt.scatter(x, y, s=1)
plt.plot([0, 1], [1, 0], c='r')
plt.title("Scatter: x+y<= 1, x>=0; y>=0")
plt.xlabel('x')
plt.ylabel('y')
```
